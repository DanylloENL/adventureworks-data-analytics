```python
import pandas as pd
import sqlite3
import requests
import io

#-------------------------------
# 1) URLs dos CSVs oficiais
#-------------------------------

urls = {
    "Calendar": "https://raw.githubusercontent.com/cnduta/Azure_AdventureWorks_DE_pipeline/main/data/Raw/AdventureWorks_Calendar.csv",
    "Customers": "https://raw.githubusercontent.com/cnduta/Azure_AdventureWorks_DE_pipeline/main/data/Raw/AdventureWorks_Customers.csv",
    "Products": "https://raw.githubusercontent.com/cnduta/Azure_AdventureWorks_DE_pipeline/main/data/Raw/AdventureWorks_Products.csv",
    "Sales_2015": "https://raw.githubusercontent.com/cnduta/Azure_AdventureWorks_DE_pipeline/main/data/Raw/AdventureWorks_Sales_2015.csv",
    "Sales_2016": "https://raw.githubusercontent.com/cnduta/Azure_AdventureWorks_DE_pipeline/main/data/Raw/AdventureWorks_Sales_2016.csv",
    "Sales_2017": "https://raw.githubusercontent.com/cnduta/Azure_AdventureWorks_DE_pipeline/main/data/Raw/AdventureWorks_Sales_2017.csv"
}

#-------------------------------
# 2) Criando bando SQLite
#-------------------------------

conn = sqlite3.connect("adventureworks.db")

print("Baixando e importando tabelas...")

for table, url in urls.items():
    print(f"Importando {table}...")
    response = requests.get(url)
    df = pd.read_csv(io.StringIO(response.text))
    df.to_sql(table, conn, if_exists="replace", index=False)

conn.close()

print("Banco adventuresworks.db criado com sucesso!")
```

## 4. Primeiras Queries

### 4.1 Total de vendas por ano

```python
def run(query):
    cursor.execute(query)
    rows = cursor.fetchall()
    for row in rows:
        print(row)
```

```python
run("""
SELECT
    strftime('%Y', OrderDate) AS Ano,
    SUM(OrderQuantity) AS Total_Qtd
FROM Sales_All
GROUP BY Ano
ORDER BY Ano;
""")
```

#### Porque o resultado deu "(None, 84174)"?

Significa que:
* Ano = None
* Tota_Qtd = 84174

Ou seja:

O SQLite não conseguiu extrair o ano da coluna "OrderDate", e por isso retornou None.
Mas conseguiu fazer a soma "SUM(OrderQuantity) e retornou 84.174 vendas.

Portanto:
✔ O banco está funcionando
✔ A view Sales_All está ok
✔ Os dados foram importados
❌ O formato da data não está reconhecido como data para o strftime()
Isso acontece muito quando o CSV tem datas no formato diferente do esperado.

##### Investigando a causa

Os CSVs do AdventureWorks (dependendo do repositório) podem vir assim:

* 1/1/2015
* 2015-01-01 00:00:00
* 2015-01-01
* 2015/01/01

Então primeiro precisamos ver um exemplo real da coluna OrderDate.

```python
run("""
SELECT OrderDate
FROM Sales_ALL
LIMIT 10;
""")
```

##### Solução: Converter o formato da data

A coluna OrderDate está no formato M/D/YYYY, que é um formato texto e não é reconhecido pelo SQLite como data nativa.

SQLite só consegue usar strftime() quando a coluna está em um formato aceito como data:

 ✔ YYYY-MM-DD

 ✔ YYYY-MM-DD HH:MM:SS

 ✔ Formatos ISO

Esse formato M/D/YYYY não segue o padrão ISO, então o SQLite interpreta como texto comum, não data.

```python
run("""
SELECT
    substr(OrderDate, -4) AS Ano,
    SUM(OrderQuantity) AS Total_Qtd
FROM Sales_All
GROUP BY Ano
ORDER By Ano;
""")
```

##### Por que usar substr(OrderDate, -4)?

Porque:

O SQLite lê a data como texto
-4 pega os 4 últimos caracteres
Como o ano sempre tem 4 dígitos → funciona para qualquer M/D/YYYY

É uma técnica comum em ETL leve quando não queremos ainda transformar toda a coluna em data nativa.

### 4.2 Top 10 Produtos mais Vendidos

```python
run("""
SELECT
    p.ProductKey,
    p.ProductName,
    SUM(s.OrderQuantity) AS TotalVendida
FROM Sales_All s
JOIN Products p ON p.ProductKey = s.ProductKey
GROUP BY p.ProductKey, p.ProductName
ORDER BY TotalVendida DESC
LIMIT 10;
""")
```

### 4.3 Top 10 Clientes (por quantidade comprada)

```python
run("""
SELECT
  c.CustomerKey,
  c.FirstName || ' ' || c.LastName AS FullName,
  SUM(s.OrderQuantity) AS TotalComprado
FROM Sales_All s
JOIN Customers c ON c.CustomerKey = s.CustomerKey
GROUP BY c.CustomerKey, FullName
ORDER BY TotalComprado DESC
LIMIT 10;
""")
```

### 4.4 Sazonalidade das vendas (por mês)

```python
run("""
SELECT
  substr(OrderDate, 1, instr(OrderDate, '/') - 1) AS Mes,
  SUM(OrderQuantity) AS TotalVendida
FROM Sales_All
GROUP BY Mes
ORDER BY CAST(Mes AS INT);
""")
```

# 1. Ler os dados do SQLite

```python
import sqlite3
import pandas as pd

# Conectar ao banco
conn = sqlite3.connect("adventureworks.db")

# Carregar tabelas
sales = pd.read_sql("SELECT * FROM Sales_All", conn)
products = pd.read_sql("SELECT * FROM Products", conn)
customers = pd.read_sql("SELECT * FROM Customers", conn)

conn.close()
```

## 1.1 Normalizar datas (regra críica de ETL)

Atualmente a data está assim:

**1/3/2015**

Isso NÃO pode ir para o Power BI desse jeito.

### Converter para formato ISO (YYYY-MM-DD):

```python
sales['OrderDate'] = pd.to_datetime(sales['OrderDate'], format='%m/%d/%Y')
```

### Agora podemos criar colunas derivadas: 

```python
sales['Year'] = sales['OrderDate'].dt.year
sales['Month'] = sales['OrderDate'].dt.month
```

## 1.2 Criar Dimensão cliente (nome completo)

No banco, nomes estavam separados.

Agora criamos isso no ETL, onde faz mais sentido.

```python
customers['FullName'] = customers['FirstName'] + ' ' + customers['LastName']
```

## 1.3 Juntando tudo (criar fato consolidada)

Criando a **Fato Vendas**, como um Data Warehouse real. 

```python
fact_sales = (
    sales
    .merge(products, on="ProductKey", how="left")
    .merge(customers[['CustomerKey','FullName']], on="CustomerKey", how="left")
)
```

### Agora temos:

* Data
* Produto
* Cliente
* Quantidade
* Tudo num único dataset.

## 1.4 Criando métricas de negócio 

#### Vamos criar um preço médio por produto: 

```python
import numpy as np

np.random.seed(42)
fact_sales['UnitPrice'] = np.random.uniform(5, 100, size=len(fact_sales))
```

### Criar métricas:

```python
fact_sales['TotalValue'] = fact_sales['OrderQuantity'] * fact_sales['UnitPrice']
fact_sales['Cost'] = fact_sales['TotalValue'] * np.random.uniform(0.5, 0.8)
fact_sales['GrossMargin'] = fact_sales['TotalValue'] - fact_sales['Cost']
```

### Isso permite: 

* Receita
* Custo
* Margem
* KPIs financeiros no Power BI

## 1.5 Selecionando colunas finais (pensando em BI)

```python
final_dataset = fact_sales[
    [
        'OrderDate',
        'Year',
        'Month',
        'FullName',
        'ProductName',
        'OrderQuantity',
        'UnitPrice',
        'TotalValue',
        'Cost',
        'GrossMargin'
    ]
].copy()
```

## 1.6 Exportar para o Power BI

```python
final_dataset.to_csv("fact_sales_powerbi.csv", index=False)
```

## BLOCO 1 - Criando métricas

O princípio aqui é:

💡 Tudo que for cálculo → Python

💡 Tudo que for visual → Power BI Web

```python
final_dataset.head(5)
```

```python
## ✅ Métrica 1 — Margem percentual

final_dataset['MarginPct'] = final_dataset['GrossMargin'] / final_dataset['TotalValue']
```

```python
## ✅ Métrica 2 — Vendas agregadas por Ano (para KPIs e gráfico)

sales_by_year = (
    final_dataset
    .groupby('Year', as_index=False)
    .agg(
        TotalSales=('TotalValue', 'sum'),
        TotalQuantity=('OrderQuantity', 'sum'),
        TotalMargin=('GrossMargin', 'sum')  
    )
)

sales_by_year['MarginPct'] = sales_by_year['TotalMargin'] / sales_by_year['TotalSales']
```

```python
## ✅ Métrica 3 — Top 10 Produtos

top_products = (
    final_dataset
    .groupby('ProductName', as_index=False)
    .agg(
        TotalQuantity=('OrderQuantity', 'sum'),
        TotalSales=('TotalValue', 'sum')
    
    )
    .sort_values('TotalQuantity', ascending=False)
    .head(10)
)
```

```python
## ✅ Métrica 4 — Top 10 Clientes

top_customers = (
    final_dataset
    .groupby('FullName', as_index=False)
    .agg(
        TotalQuantity=('OrderQuantity', 'sum'),
        TotalSales=('TotalValue', 'sum')
    )
    .sort_values('TotalQuantity', ascending=False)
    .head(10)
)
```

```python
## ✅ Métrica 5 — Sazonalidade (por mês) 

sales_by_month = (
    final_dataset
    .groupby('Month', as_index=False)
    .agg(
        TotalQuantity=('OrderQuantity', 'sum'),
        TotalSales=('TotalValue', 'sum')
    )
    .sort_values('Month')
)
```

## BLOCO 2 - Exportação dos arquivos CSV

```python
final_dataset.to_csv("fact_sales_powerbi.csv", index=False)
sales_by_year.to_csv("kpi_sales_by_year.csv", index=False)
top_products.to_csv("top_products.csv", index=False)
top_customers.to_csv("top_customers.csv", index=False)
sales_by_month.to_csv("sales_by_month.csv", index=False)
```
